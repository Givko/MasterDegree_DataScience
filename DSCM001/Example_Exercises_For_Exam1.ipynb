```python
import numpy as np

A = np.random.rand(1500, 1500)
B = np.random.rand(1500, 1500)

left = (A @ B).T
right = B.T @ A.T

print("AT*BT = (AB)T" if np.allclose(left, right) else "Not equal")

```

```python
det_AB = np.linalg.det(A @ B)
det_A_multi_B = np.linalg.det(A) * np.linalg.det(B)

print("|A*B| = |A|*|B|" if np.equal(det_AB, det_A_multi_B) else "Not equal")
```

```python
det_inverse_A = np.linalg.det(np.linalg.inv(A))

det_A = np.linalg.det(A)

print("|A*B| = |A|*|B|" if np.isclose(det_inverse_A, 1/det_A) else "Not equal")
```

```python
def check_symmetric(A) -> bool:
  return np.allclose(A, A.T)

def is_matrix_positive_definite(A, eigvalues):
  """
    Check if the matrix is symmetric and negative definite.

    Args:
        A: Input square matrix.

    Returns:
        A tuple (is_positive_definite, is_semi_positive_definite):
        - is_positive_definite
        - is_semi_positive_definite
    """

  is_symm = check_symmetric(A)
  are_eigvalues_positive = np.all(eigvalues > 0)
  are_eigvalues_zero_or_positive = np.all(eigvalues >= 0)

  is_positive_definite = is_symm and are_eigvalues_positive
  is_semi_positive_definite = is_symm and are_eigvalues_zero_or_positive and not is_positive_definite

  return is_positive_definite, is_semi_positive_definite

def is_matrix_negative_definite(A, eigvalues):
  """
    Check if the matrix is symmetric and negative definite.

    Args:
        A: Input square matrix.

    Returns:
        A tuple (is_negative_definite, is_semi_negative_definite):
        - is_negative_definite
        - is_semi_negative_definite
    """

  is_symm = check_symmetric(A)
  are_eigvalues_negative = np.all(eigvalues < 0)
  print(eigvalues)
  print(are_eigvalues_negative)
  are_eigvalues_zero_or_negative = np.all(eigvalues <= 0)

  is_negative_definite = is_symm and are_eigvalues_negative
  is_semi_negative_definite = is_symm and are_eigvalues_zero_or_negative and not is_negative_definite

  return is_negative_definite, is_semi_negative_definite

A = np.array([[2,-1,-1],[-1,2,-1],[-1,-1,2]])
B = np.array([[2,-1,-1],[-1,2,1],[-1,1,2]])
C = np.array([[1,2,3],[2,5,4],[3,4,9]])
D = np.array([[1,2,0,0],[2,6,-2,0],[0,-2,5,-2],[0,0,-2,3]])

A_eigvalues = np.linalg.eigvals(A)
B_eigvalues = np.linalg.eigvals(B)
C_eigvalues = np.linalg.eigvals(C)
D_eigvalues = np.linalg.eigvals(D)

is_A_pos_def, is_A_semi_pos_def = is_matrix_positive_definite(A, A_eigvalues);
is_B_pos_def, is_B_semi_pos_def = is_matrix_positive_definite(B, B_eigvalues);
is_C_pos_def, is_C_semi_pos_def = is_matrix_positive_definite(C, C_eigvalues);
is_D_pos_def, is_D_semi_pos_def = is_matrix_positive_definite(D, D_eigvalues);

print("A is pos def" if is_A_pos_def else "Not pos def")
print("B is pos def" if is_B_pos_def else "Not pos def")
print("C is pos def" if is_C_pos_def else "Not pos def")
print("D is pos def" if is_D_pos_def else "Not pos def")

print("A is semi pos def" if is_A_semi_pos_def else "Not semi pos def")
print("B is semi pos def" if is_B_semi_pos_def else "Not semi pos def")
print("C is semi pos def" if is_C_semi_pos_def else "Not semi pos def")
print("D is semi pos def" if is_D_semi_pos_def else "Not semi pos def")

is_A_neg_def, is_A_semi_neg_def = is_matrix_negative_definite(A, A_eigvalues);
print(is_A_neg_def, is_A_semi_neg_def)
# is_B_neg_def, is_B_semi_neg_def = is_matrix_negative_definite(B, B_eigvalues);
# is_C_neg_def, is_C_semi_neg_def = is_matrix_negative_definite(C, C_eigvalues);
# is_D_neg_def, is_D_semi_neg_def = is_matrix_negative_definite(D, D_eigvalues);

# print("A is neg def" if is_A_neg_def else "Not neg def")
# print("B is neg def" if is_B_neg_def else "Not neg def")
# print("C is neg def" if is_C_neg_def else "Not neg def")
# print("D is neg def" if is_D_neg_def else "Not neg def")

# print("A is semi neg def" if is_A_semi_neg_def else "Not semi neg def")
# print("B is semi neg def" if is_B_semi_neg_def else "Not semi neg def")
# print("C is semi neg def" if is_C_semi_neg_def else "Not semi neg def")
# print("D is semi neg def" if is_D_semi_neg_def else "Not semi neg def")
```
